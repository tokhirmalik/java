```python
%%writefile Main.java
public class Main {
    public static void main(String[] args) {

        int n = 31;   // nechta Fibonacci soni kerak
        int a = 0, b = 1;

        int count = 1;

        System.out.println("Fibonacci numbers:");

        do {
            System.out.print(a + " ");
            int next = a + b;
            a = b;
            b = next;
            count++;
        } while (count <= n);

    }
}
```

```python
%%shell
javac Main.java
```
