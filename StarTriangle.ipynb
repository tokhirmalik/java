```python
%%writefile StarTriangle.java
public class StarTriangle {
    public static void main(String[] args) {

        int rows = 4; // nechta qator kerak

        for (int i = 1; i <= rows; i++) {   // qatorlar
            for (int j = 1; j <= i; j++) {  // har qatorda i ta yulduzcha
                System.out.print(" *");
            }
            System.out.println();           // keyingi qatorga o'tish
        }
    }
}
```

```python
%%shell
javac StarTriangle.java
```
