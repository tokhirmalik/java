```python
%%writefile Bank.java
import java.util.Scanner;

public class Bank {

    static void display(String name, long accNo, double balance) {
        System.out.println("\n--- ACCOUNT DETAILS ---");
        System.out.println("Name: " + name);
        System.out.println("Account Number: " + accNo);
        System.out.printf("Balance: %.2f%n", balance);
    }

    static double deposit(double balance, double amount) {
        if (amount <= 0) {
            System.out.println("Amount must be positive!");
            return balance;
        }
        balance += amount;
        System.out.println("Deposit successful.");
        return balance;
    }

    static double withdraw(double balance, double amount) {
        if (amount <= 0) {
            System.out.println("Amount must be positive!");
            return balance;
        }
        if (amount > balance) {
            System.out.println("Insufficient funds!");
            return balance;
        }
        balance -= amount;
        System.out.println("Withdrawal successful.");
        return balance;
    }

    static void showBalance(double balance) {
        System.out.printf("Your balance: %.2f%n", balance);
    }

    public static void main(String[] args) {

        Scanner sc = new Scanner(System.in);

        System.out.print("Enter account owner name: ");
        String name = sc.nextLine();

        long accNo = 0;
        while (true) {
            System.out.print("Enter account number: ");
            try {
                accNo = Long.parseLong(sc.nextLine());
                break;
            } catch (Exception e) {
                System.out.println("Numbers only! Try again.");
            }
        }

        double balance = 3122.0;

        while (true) {
            System.out.println("\n--- MENU ---");
            System.out.println("1. Account Details");
            System.out.println("2. Deposit");
            System.out.println("3. Withdraw");
            System.out.println("4. Check Balance");
            System.out.println("5. Exit");
            System.out.print("Choose: ");

            int choice;
            try {
                choice = Integer.parseInt(sc.nextLine());
            } catch (Exception e) {
                System.out.println("Invalid input. Enter number 1–5.");
                continue;
            }

            switch (choice) {
                case 1:
                    display(name, accNo, balance);
                    break;

                case 2:
                    System.out.print("Enter deposit amount: ");
                    try {
                        balance = deposit(balance, Double.parseDouble(sc.nextLine()));
                    } catch (Exception e) {
                        System.out.println("Invalid amount.");
                    }
                    break;

                case 3:
                    System.out.print("Enter withdraw amount: ");
                    try {
                        balance = withdraw(balance, Double.parseDouble(sc.nextLine()));
                    } catch (Exception e) {
                        System.out.println("Invalid amount.");
                    }
                    break;

                case 4:
                    showBalance(balance);
                    break;

                case 5:
                    System.out.println("Goodbye!");
                    sc.close();
                    return;

                default:
                    System.out.println("Choose between 1–5!");
            }
        }
    }
}
```
