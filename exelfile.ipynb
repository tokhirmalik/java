```python
%%writefile Main.java
import java.util.Scanner;
import java.io.File;

public class Main {
    public static void main(String[] args) {
        File file = new File("exel.csv");
        try {
            Scanner sc = new Scanner(file);
            sc.nextLine();  // Skip header line

            int count = 0;
            int sum = 0;
            int max = Integer.MIN_VALUE;
            int min = Integer.MAX_VALUE;
            int pass = 0;
            int fail = 0;

            while (sc.hasNextLine()) {
                String line = sc.nextLine();
                String[] parts = line.split(",");

                String name = parts[0];
                int marks = Integer.parseInt(parts[1]);

                System.out.println(name + " ----> " + marks);

                count++;
                sum += marks;
                if (marks > max) max = marks;
                if (marks < min) min = marks;

                if (marks >= 40) pass++;
                else fail++;
            }
            sc.close();

            System.out.println("**** Student Analysis Report ****");
            System.out.println("Total students: " + count);
            System.out.println("Highest Mark: " + max);
            System.out.println("Lowest Mark: " + min);
            System.out.println("No of students pass: " + pass);
            System.out.println("No of students fail: " + fail);

        } catch (Exception e) {
            System.out.println("File not found!!!!");
        }
    }
}
```
