```python
%%writefile Main.java
import java.io.BufferedReader;
import java.io.FileReader;

public class Main {

    public static void main(String[] args) {
        rainydays();
        mintemperature();
        windspeedmarch();
    }

    public static void rainydays() {
        int count = 0;
        try (BufferedReader br = new BufferedReader(new FileReader("weather.csv"))) {

            br.readLine(); // skip header

            String line;
            while ((line = br.readLine()) != null) {
                String[] data = line.split(",");
                double humidity = Double.parseDouble(data[2]);
                if (humidity > 70) count++;
            }

        } catch (Exception e) {
            System.out.println("File not Found!!");
        }

        System.out.println("Number of RainyDays: " + count);
    }

    public static void mintemperature() {
        double mintemp = Double.MAX_VALUE;
        String date = "";
        String city = "";
        try (BufferedReader br = new BufferedReader(new FileReader("weather.csv"))) {

            br.readLine(); // skip header

            String line;
            while ((line = br.readLine()) != null) {
                String[] data = line.split(",");
                double temp = Double.parseDouble(data[1]);
                if (temp < mintemp) {
                    mintemp = temp;
                    date = data[0];
                    city = data[4];
                }
            }

        } catch (Exception e) {
            System.out.println("File not Found!!");
        }

        System.out.println("Minimum Temperature is: " + mintemp + " on " + date + " in " + city);
    }

    public static void windspeedmarch() {
        int count = 0;
        double total = 0;
        try (BufferedReader br = new BufferedReader(new FileReader("weather.csv"))) {

            br.readLine(); // skip header

            String line;
            while ((line = br.readLine()) != null) {
                String[] data = line.split(",");
                String date = data[0];
                if (date.startsWith("2025-03")) {
                    double wind = Double.parseDouble(data[3]);
                    total += wind;
                    count++;
                }
            }

        } catch (Exception e) {
            System.out.println("File not Found!!");
        }

        double mean = (count > 0) ? total / count : 0;
        System.out.println("Average wind speed in March is: " + mean + " and " + count + " records");
    }
}
```
