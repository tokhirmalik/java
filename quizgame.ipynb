```python
%%writefile Main.java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {
        String[] questions = {
            "What does CPU stand for?",
            "Which device is used to display output?",
            "What is the brain of the computer?",
            "Which of the following is an input device?",
            "What type of memory is temporary?",
            "Which of these is an example of an operating system?",
            "What does 'www' stand for in a website address?",
            "Which key is used to delete text to the left of the cursor?",
            "Which type of memory is permanent?",
            "Which company developed the Windows operating system?"
        };

        String[] answers = {
            "centralprocessingunit",
            "monitor",
            "cpu",
            "keyboard",
            "ram",
            "windows",
            "worldwideweb",
            "backspace",
            "harddisk",
            "microsoft"
        };

        Scanner abc = new Scanner(System.in);
        String choice;

        do {
            int score = 0;

            for (int i = 0; i < questions.length; i++) {
                System.out.println("Question " + (i + 1) + ": " + questions[i]);
                System.out.print("Your answer: ");
                String ans = abc.next().toLowerCase();

                if (ans.equals(answers[i])) {
                    System.out.println("✅ Correct answer!");
                    score++;
                } else {
                    System.out.println("❌ Wrong answer! The correct answer is: " + answers[i]);
                }
                System.out.println();
            }

            System.out.println("Your total score is: " + score + "/" + questions.length);
            System.out.print("Do you want to continue? (yes/no): ");
            choice = abc.next();

        } while (choice.equalsIgnoreCase("yes"));

        abc.close();
    }
}
```
